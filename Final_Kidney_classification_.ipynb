```python
import shutil
import os

base_path = "temp_data/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone"

# Destination banao
os.makedirs("artifacts/data_ingestion/kidney-ct-scan-image/Normal", exist_ok=True)
os.makedirs("artifacts/data_ingestion/kidney-ct-scan-image/Tumor", exist_ok=True)

# Copy karo
shutil.copytree(f"{base_path}/Normal",
                "artifacts/data_ingestion/kidney-ct-scan-image/Normal",
                dirs_exist_ok=True)

shutil.copytree(f"{base_path}/Tumor",
                "artifacts/data_ingestion/kidney-ct-scan-image/Tumor",
                dirs_exist_ok=True)

# Count karo
normal_count = len(os.listdir("artifacts/data_ingestion/kidney-ct-scan-image/Normal"))
tumor_count = len(os.listdir("artifacts/data_ingestion/kidney-ct-scan-image/Tumor"))
print(f"Normal images: {normal_count}")
print(f"Tumor images: {tumor_count}")
print("Dataset ready!")

# Cleanup
shutil.rmtree("temp_data")
!rm ct-kidney-dataset-normal-cyst-tumor-and-stone.zip
print("Cleanup done!")
```
